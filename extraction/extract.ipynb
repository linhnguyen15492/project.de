```python
from datetime import datetime


def is_datetime_format(date_string, date_format):
    try:
        datetime.strptime(date_string, date_format)
        return True
    except ValueError:
        return False


# Example with a valid datetime string
string1 = "2023-10-27 15:30:00"
format1 = "%Y-%m-%d %H:%M:%S"
print(f"'{string1}' is in '{format1}' format: {is_datetime_format(string1, format1)}")

# Example with an invalid datetime string for the given format
string2 = "October 27, 2023"
format2 = "%Y-%m-%d %H:%M:%S"
print(f"'{string2}' is in '{format2}' format: {is_datetime_format(string2, format2)}")

# Example with a different valid format
string3 = "27/10/2023"
format3 = "%d/%m/%Y"
print(f"'{string3}' is in '{format3}' format: {is_datetime_format(string3, format3)}")
```

```python
from configparser import ConfigParser


def config(filename="database.ini", section="postgresql"):
    # create a parser
    parser = ConfigParser()
    # read config file
    parser.read(filename)

    # get section, default to postgresql
    db = {}
    if parser.has_section(section):
        params = parser.items(section)
        for param in params:
            db[param[0]] = param[1]
    else:
        raise Exception(
            "Section {0} not found in the {1} file".format(section, filename)
        )

    return db
```

```python
import psycopg2
from psycopg2 import extras
from datetime import datetime, date


def bulk_insert_execute_batch(sql, data_list):
    conn = None
    try:
        conn = psycopg2.connect(
            host="localhost",
            database="bootcamp",
            user="postgres",
            password="123456",
            port="5432",  # or your specific port
        )
        cur = conn.cursor()

        sql = sql
        extras.execute_batch(cur, sql, data_list)
        conn.commit()
        print(f"{len(data_list)} rows inserted successfully using execute_batch.")

    except (Exception, psycopg2.Error) as error:
        print(f"Error: {error}")
    finally:
        if conn:
            cur.close()
            conn.close()
```

```python
def extract(path: str, sheetname: str, mappings: dict) -> pd.DataFrame:
    # read excel file
    try:
        df = pd.read_excel(path, sheet_name=sheetname)

        # add column to note datasource
        filename = str(path.split("\\")[-1].strip())
        df["reference_source"] = filename

        # add new column to mappings object
        mappings["reference_source"] = "reference_source"

        # rename columns to matching with database
        df.rename(columns=mappings, inplace=True)

        return df[[col for col in mappings.values()]]

    except Exception as e:
        print(f"Fail to read {path} with sheet {sheetname}")

    return pd.DataFrame()
```

```python
def convert_datetime_to_date(
    df: pd.DataFrame, columns: list, date_format: str
) -> pd.DataFrame:
    for column in columns:
        try:
            df[column] = pd.to_datetime(df[column], format=date_format).dt.date
            print(f"Convert column {column} to date successfully")
        except:
            print(f"Fail to convert {column} from datetime to date")

    return df
```

```python
def data_loader(data: pd.DataFrame, schema, table) -> bool:
    columns_insert = list(data.columns)
    sql = f"INSERT INTO {schema}.{table} ({', '.join(columns_insert)}) VALUES ({', '.join(['%s'] * len(columns_insert))})"
    data_to_insert = [tuple(x) for x in data.to_numpy()]
    try:
        bulk_insert_execute_batch(sql, data_to_insert)
        return True
    except Exception as e:
        print(e)
    return False
```
